```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
txt = open('./data/song.txt').read()
mask_img = np.array(Image.open('./data/mk.jpg'))

wc = WordCloud(width=800, 
               height=800, 
               mask=mask_img,
               background_color='white', 
               font_path='./data/H2MJRE.TTF').generate(txt)

plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
