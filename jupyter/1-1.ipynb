```python
import matplotlib.pyplot as plt

plt.title('Plot4')
plt.plot(['a','b','c','d','e'],
         [4,7,3,8,3],
         c='orange',
         ls='--',
         marker='o')
plt.show()
```

```python
# x축 y축 범위설정
plt.title('Plot4')
plt.plot(['a','b','c','d','e'],
         [4,7,3,8,3],
         c='orange',
         ls=':',
         marker='o')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.show()
```

```python
# 여러개 라인 그리기
import numpy as np

# 0에서 5까지, 간격은 0.2로
t = np.arange(0, 5, 0.2)
t
```

```python
plt.title('Plot5')
plt.plot(t, t, 'r--')
plt.plot(t, t**2, 'b--')
plt.plot(t, t**3, 'g--')
# plot(x축값, y축값, 디자인)
```
