```python
# 네이버 실검(몽고db) to 워드클라우드

# 몽고db 접속
from pymongo import MongoClient as mongo
conn = mongo('mongodb://hyh:1234@192.168.111.101:27017')
db = conn.get_database('test')
naver = db.get_collection('NAVER')

# 워드클라우드
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
results = naver.find()
txt = ''
for result in results:
    txt += result['1']+','+result['2']+','+result['3']+','+result['4']+','+result['5']+','+result['6']+','+result['7']+','+result['8']+','+result['9']+','+result['10']+','
txt
```

```python
mask_img = np.array(Image.open('./data/muji.jpg'))
wc = WordCloud(width=800, 
               height=800, 
               mask=mask_img, 
               background_color='white', 
               font_path='./data/H2MJRE.TTF').generate(txt)

plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
