```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rc('font', family='nanumgothic')
mpl.rc('axes', unicode_minus=False)
```

```python
df_exam = pd.read_csv('./data/exam.csv')

df_exam.plot()
plt.title('df_exam 그래프')
```

```python
# 수학컬럼 막대그래프

plt.bar(df_exam.id, df_exam.math)
```

```python
# 영어 막대그래프

plt.bar(df_exam.id, df_exam.english)
```

```python
# 평균 히스토그램
df_exam['total'] = df_exam.math + df_exam.english + df_exam.science
df_exam['mean'] = df_exam.total / 3

df_exam['mean'].hist()
```
