```python
import matplotlib as mpl
import matplotlib.pylab as plt
import numpy as np
```

```python
# 한글 사용하기
mpl.rc('font', family='nanumgothic')
mpl.rc('axes', unicode_minus=False)
```

```python
# 바 차트
x = [1,2,3]
y = [2,3,1]

plt.title('Bar Chart')
plt.xticks(x, ['가','나','다'])
plt.ylabel('y라벨')
plt.bar(x,y)
```

```python
# 히스토그램(분포도)

# -3~+3까지의 임의 정수, 산모양 분포도의 가우시안 표준 정규분포도
data = np.random.randn(1000)
data.round()

plt.title('정규분포도')
plt.hist(data)
```

```python
# 0에서 1 사이의 랜덤값 1000개, 균일한 분포도
x1 = np.random.rand(1000)
x1.round()

plt.title('균일한 분포도')
plt.hist(x1)
```

```python
# scatter, 산포도, 그래프를 점으로 표시한 것
x = np.random.randn(1000)
y = data
plt.title('산포도')
plt.scatter(x, y)
```
